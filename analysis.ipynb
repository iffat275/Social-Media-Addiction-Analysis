```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dono files load karo
df1 = pd.read_csv(r"D:\Imp docu\Social Media Addiction Analysis\Datasets\screen_time_behavior.csv")
df2 = pd.read_csv(r"D:\Imp docu\Social Media Addiction Analysis\Datasets\tiktok_instagram_global_100countries.csv")

print("File 1 shape:", df1.shape)
print("File 2 shape:", df2.shape)

df1.head()
```

```python
#Basic info about the dataset
print("Shape:", df1.shape)
print("\nColumns:", df1.columns.tolist())
print("\nMissing Values:\n", df1.isnull().sum())
print("\nBasic Stats:\n", df1.describe())
```

```python
print("File 2 Columns:", df2.columns.tolist())
print("\nMissing Values df1:\n", df1.isnull().sum())
print("\nMissing Values df2:\n", df2.isnull().sum())
```

```python
# Platform wise usage
plt.figure(figsize=(10,5))
sns.countplot(x="platform", data=df1)
plt.title("Most Used Platforms")
plt.xlabel("Platform")
plt.ylabel("Users")
plt.show()
```

```python
# Age group wise screen time
plt.figure(figsize=(10,5))
sns.barplot(x="age_group", y="weekday_screen_hours", data=df1)
plt.title("Age Group vs Weekday Screen Time")
plt.xlabel("Age Group")
plt.ylabel("Screen Hours")
plt.show()
```

```python
# Gender wise screen time comparison
plt.figure(figsize=(10,5))
sns.barplot(x="gender", y="weekday_screen_hours", data=df1, palette="pastel")
plt.title("Gender vs Weekday Screen Time")
plt.xlabel("Gender")
plt.ylabel("Screen Hours")
plt.show()
```

```python
# Weekday vs Weekend comparison
plt.figure(figsize=(10,5))
sns.barplot(x="platform", y="weekday_screen_hours", data=df1, label="Weekday", color="blue", alpha=0.6)
sns.barplot(x="platform", y="weekend_screen_hours", data=df1, label="Weekend", color="orange", alpha=0.6)
plt.title("Weekday vs Weekend Screen Time by Platform")
plt.xlabel("Platform")
plt.ylabel("Screen Hours")
plt.legend()
plt.show()
```

```python
# df2 columns dekho pehle
df2.head()
```

```python
# Country wise addiction level
plt.figure(figsize=(12,5))
sns.countplot(x="addiction_level", data=df2, palette="pink")
plt.title("Addiction Level Distribution")
plt.xlabel("Addiction Level")
plt.ylabel("Count")
plt.show()
```

```python
# Top 10 countries by average addiction score
plt.figure(figsize=(12,6))
top_countries = df2.groupby("country")["addiction_score"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Reds")
plt.title("Top 10 Countries by Addiction Score")
plt.xlabel("Avg Addiction Score")
plt.ylabel("Country")
plt.show()
```

```python
# Sleep vs Addiction
plt.figure(figsize=(10,5))
sns.scatterplot(x="sleep_hours", y="addiction_score", data=df2, hue="addiction_level", palette="Reds")
plt.title("Sleep Hours vs Addiction Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Addiction Score")
plt.show()
```

```python
# TikTok vs Instagram comparison
plt.figure(figsize=(10,5))
avg_usage = pd.DataFrame({
    'Platform': ['TikTok', 'Instagram'],
    'Avg Minutes': [df2['tiktok_minutes_daily'].mean(), df2['instagram_minutes_daily'].mean()]
})
sns.barplot(x='Platform', y='Avg Minutes', data=avg_usage, palette='Reds')
plt.title("TikTok vs Instagram - Average Daily Usage")
plt.xlabel("Platform")
plt.ylabel("Average Minutes Per Day")
plt.show()
```

```python
# Age vs Addiction Score
plt.figure(figsize=(10,5))
sns.scatterplot(x="age", y="addiction_score", 
                data=df2, hue="addiction_level", palette="Reds")
plt.title("Age vs Addiction Score")
plt.xlabel("Age")
plt.ylabel("Addiction Score")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
numeric_cols = df2.select_dtypes(include='number')
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".1f", cmap="Reds")
plt.title("Correlation Heatmap - Social Media Addiction")
plt.show()
```

```python
# Sleep hours vs Mental Health Support
plt.figure(figsize=(10,5))
sns.scatterplot(x="sleep_hours", y="mental_health_support_index", 
                data=df2, hue="addiction_level", palette="Reds")
plt.title("Sleep Hours vs Mental Health Support Index")
plt.xlabel("Sleep Hours")
plt.ylabel("Mental Health Support Index")
plt.show()
```

```python
# Country wise average sleep hours - Bottom 10 (least sleep)
plt.figure(figsize=(12,6))
sleep_countries = df2.groupby("country")["sleep_hours"].mean().sort_values().head(10)
sns.barplot(x=sleep_countries.values, y=sleep_countries.index, palette="Blues")
plt.title("Top 10 Countries with Least Sleep Hours")
plt.xlabel("Avg Sleep Hours")
plt.ylabel("Country")
plt.show()
```

```python
# Impulsivity vs Addiction Score
plt.figure(figsize=(10,5))
sns.scatterplot(x="impulsivity_index", y="addiction_score", 
                data=df2, hue="addiction_level", palette="Reds")
plt.title("Impulsivity vs Addiction Score")
plt.xlabel("Impulsivity Index")
plt.ylabel("Addiction Score")
plt.show()
```

```python
# Dopamine Dependency vs Addiction Score
plt.figure(figsize=(10,5))
sns.scatterplot(x="dopamine_dependency_score", y="addiction_score",
                data=df2, hue="addiction_level", palette="Reds")
plt.title("Dopamine Dependency vs Addiction Score")
plt.xlabel("Dopamine Dependency Score")
plt.ylabel("Addiction Score")
plt.show()
```
